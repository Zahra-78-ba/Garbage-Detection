```
import kagglehub
viswaprakash1990_garbage_detection_path = kagglehub.dataset_download('viswaprakash1990/garbage-detection')
```

```
import os
print(os.listdir(viswaprakash1990_garbage_detection_path))
```

```
data_yaml=viswaprakash1990_garbage_detection_path+'/data.yaml'
print(data_yaml)
```

```
import ultralytics
from ultralytics import YOLO
data_yaml = viswaprakash1990_garbage_detection_path + '/GARBAGE CLASSIFICATION/data.yaml'
model=YOLO("yolov8n.pt")
model.train(
    data=data_yaml,
    epochs=30,
    imgsz=640,
    batch=8
)
```

```
best_model_path="runs/detect/train/weights/best.pt"
print(best_model_path)
```
